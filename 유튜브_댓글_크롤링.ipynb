```python
#os.chdir("D:/myfriands10_google/Mixture_lab/youtube_산출물/R결과물2")
#os.chdir("C:/Users/hyerin/Google 드라이브(myfriands10@gmail.com)/Mixture_lab/youtube_산출물/R결과물2")
os.chdir("/content/drive/MyDrive/Mixture_lab/youtube_산출물/R결과물2")
#url_dat = pd.read_csv('plur_k_kids_disaster.csv',encoding='utf-8')
#url_dat = pd.read_csv('plus_korea_covid19_kids.csv',encoding='utf-8')
url_dat = pd.read_csv('plur_usa_covid19_kid.csv',encoding='cp949')
url = url_dat.url
day = url_dat.day

url_list = [] # url 주소
day_list = [] # url 영상 업로드 날짜
comment_list = [] # 댓글
report_date_list = [] # 댓글이 업로드된 날짜(현재 날짜를 기준으로 어느시점에 기재되었는지 ex. 3개월 전)

os.chdir("../")
os.chdir("../selenium")
os.get
driver = Chrome('chromedriver.exe')
```

```python
#for i in range(len(url_dat)):
for i in range(len(url_dat)):

    driver.get(url[i])
    
    last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    for j in range(100):
        ## 페이지 아래로 내리기
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        ## 시간 간격 주기
        time.sleep(3.0)
        ## 새로운 페이지 높이
        new_page_height = driver.execute_script("return document.documentElement.scrollHeight")

        ## 더이상 안내려가면 break
        if new_page_height == last_page_height: 
            break 
        last_page_height = new_page_height
    
    ## rvest로 html구조에서 가져오는건 같은 방식
    html_source = driver.page_source
    soup = bs(html_source,'html.parser')
    soup2 = soup.find_all(attrs={'id':'content-text'})
    soup3 = soup.find_all(attrs={'class':'published-time-text'})
    
    if len(soup2)>0:
        for k in range(len(soup2)):
            url_list.append(url[i])
            day_list.append(day[i])
            comment_list.append(soup2[k].text)
            report_date_list.append(soup3[k].text)
            
    ## 댓글 없는 url에는 NA값 대입해서 한줄 만들어주기
    if len(soup2)==0:
        url_list.append(url[i])
        day_list.append(day[i])
        comment_list.append('NA')
        report_date_list.append('NA')
    
    print(f'{i+1}/{len(url_dat)} 완료')
```

```python
dat = pd.DataFrame({'url':url_list,'day':day_list,
             'comment':comment_list,'report_date':report_date_list}) 
dat
```
